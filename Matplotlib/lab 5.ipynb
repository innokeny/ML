**Сравнение значений диаметров и объемов лун и их планет**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
planets = pd.read_excel("/content/drive/MyDrive/Python/мдс/lab5/planets.xlsx")
planets
```

```python
moons = pd.read_excel("/content/drive/MyDrive/Python/мдс/lab5/moons.xlsx")
moons
```

```python
data = planets.merge(moons, how="inner", left_on="planet_name", right_on="planet_name")
data
```

```python
data = data.assign(moon_volume_km3 = 4/3 * np.pi * (data.moon_diameter_km/2)**3)
data = data.assign(planet_volume_km3= 4/3 * np.pi * (data.planet_diameter_km/2)**3)
data
```

```python
x = data["planet_diameter_km"]
y = data["moon_diameter_km"]
plt.figure(figsize=(6, 10))
sizes = np.asarray(list(x))*0.0005
markers = ["o", "v", "v", "1", "1", "1", "1", "p", "p", "p", "p", "p", "p", "p", "p", "p", "s", "s", "s", "s", "s", "h", "h"]
for xp, yp, m, sz in zip(x, y, markers, sizes):
    plt.scatter(xp, yp, marker=m, s=sz)
plt.legend((data["planet_name"]+'\n'+data["moon_name"]), bbox_to_anchor=(1.1, 1.05))
plt.xlabel('Planet diameter, km')   
plt.ylabel('Moon diameter, km')   
plt.show()
```

```python
x = data["planet_volume_km3"]
y = data["moon_volume_km3"]
plt.figure(figsize=(6, 20))
sizes = np.asarray(list(x))*0.000000000001
markers = ["o", "v", "v", "1", "1", "1", "1", "p", "p", "p", "p", "p", "p", "p", "p", "p", "s", "s", "s", "s", "s", "h", "h"]
for xp, yp, m, sz in zip(x, y, markers, sizes):
    plt.scatter(xp, yp, marker=m, s=sz)
plt.legend((data["planet_name"]+'\n'+data["moon_name"]), bbox_to_anchor=(1.1, 1))
plt.xlabel('Planet volume, km3')   
plt.ylabel('Moon volume, km3')   
plt.show()
```

```python
y = data["moon_name"]
x = data["moon_volume_km3"]
plt.figure(figsize=(10, 10))
sizes = np.asarray(list(x))*0.00000005
for xp, yp, m, sz in zip(x, y, markers, sizes):
    plt.scatter(xp, yp, marker="o", s=sz)  
plt.show()
```

**Сравнение угловых размеров солнца и лун**

```python
data = data.assign(moon_angular_diameter_arcses = 206265 * data.moon_diameter_km / data.moon_planet_distance_km)
data
```

```python
sun_diameter = 1392700
data = data.assign(sun_angular_diameter_arcses = 206265 * sun_diameter / data.planet_sun_distance_km)
data
```

```python
x = data["moon_angular_diameter_arcses"]
y = data["sun_angular_diameter_arcses"]
plt.figure(figsize=(20, 10))
sizes = np.asarray(list(x))*0.3
markers = ["o", "v", "v", "1", "1", "1", "1", "p", "p", "p", "p", "p", "p", "p", "p", "p", "s", "s", "s", "s", "s", "h", "h"]
for xp, yp, m, sz in zip(x, y, markers, sizes):
    plt.scatter(xp, yp, marker=m, s=sz)
plt.legend(('Sun'+'\n'+data["moon_name"]), bbox_to_anchor=(1.1, 1))
plt.xlabel('Moon angular diameter arcses')   
plt.ylabel('Sun angular diameter arcses')   
plt.show()
```
